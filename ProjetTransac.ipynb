```python
# Charger les DataFrames  partir des fichiers CSV
products_df = pd.read_csv('products.csv')
transactions_df = pd.read_csv('transactions.csv')

# Fusionner les deux DataFrames en utilisant la colonne 'id_prod'
transprod_df = pd.merge(transactions_df, products_df, on='id_prod', how='outer')

# Afficher le DataFrame fusionn
transprod_df
```

```python
customers_df = pd.read_csv('customers.csv')

# Fusionner les deux DataFrames en utilisant la colonne 'id_prod'
transcusto_df = pd.merge(transprod_df, customers_df, on='client_id', how='outer')

#
transcustosnan_df = transcusto_df.dropna()


# Afficher le DataFrame fusionn
print(transcustosnan_df)
```

```python
transcustosnan1_df = transcustosnan_df.sort_values(by='price').iloc[200:]

transcustosnan1_df.sort_values(by='id_prod')
```

```python
nombre_articles_par_produit = transcustosnan1_df.groupby('id_prod')['price'].count()
nombre_articles_par_produit
```

```python
median_par_categorie = transcustosnan1_df.groupby('categ')['price'].median()
median_par_categorie
```

```python
parite_hf = transcustosnan1_df['sex'].value_counts()
parite_hf.reset_index()
```

```python
pourcentage_homme_femme = parite_hf / transcustosnan1_df['sex'].count() * 100
pourcentage_homme_femme
```

```python
# Recherche de la ligne o la valeur de 'id_prod' est gale  '0_2245'
indice_colonne_id_prod = transactions_df.columns.get_loc('id_prod')
ligne_specifique = transactions_df[transactions_df.iloc[:, indice_colonne_id_prod] == '0_2245']

ligne_specifique
```

```python
client_bizare_df = transcustosnan_df.sort_values(by='price').iloc[:200].copy()
client_bizare_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Exemple de modifications des paramtres
sns.boxplot(x=transcustosnan1_df["price"], y=transcustosnan1_df["categ"],  # Utilisation de deux colonnes pour le trac
            palette="Set2",  # Utilisation d'une palette de couleurs spcifique
            orient="h",     # Orientation horizontale
            width=0.5,      # Largeur des botes
            linewidth=2.5)  # Largeur des lignes
plt.title("Rpartition des prix par catgories")
plt.xlabel("Prix")
plt.ylabel("Catgories")
plt.show()
```

```python
labonnedata_df= pd.merge(transcustosnan1_df, ligne_specifique, how='outer')
labonnedata_df
```

```python
valeurs_specifiques = {'price': 9.9, 'categ': 0.0}
labonnedata_df.fillna(value=valeurs_specifiques, inplace=True)

labonnedata_df
```

```python
labonnedata2_df = pd.merge(labonnedata_df, customers_df, on='client_id',  how='left')
labonnedata2_df
```

```python
#Correction du tableau en supprimant les mauvaise colonnes + renommer les colonnes avec le bon nom

labonnedata2_df.drop(columns=['sex_x','birth_x'], inplace=True)

labonnedata2_df.rename(columns={'sex_y' : 'sex'}, inplace=True)
labonnedata2_df.rename(columns={'birth_y' : 'birth'}, inplace=True)

labonnedata2_df
```

```python
# Crer un camembert  partir du DataFrame
pourcentage_homme_femme.plot(kind='pie', y='Pourcentage', autopct='%1.1f%%', startangle=90, colors=['red', 'green'])
plt.axis('equal')  # Assurez-vous que le camembert est un cercle
plt.title('Rpartition Hommes/Femmes')
plt.ylabel('')  # Supprimer le libell de l'axe y
plt.show()
```

```python
from datetime import datetime
# Rcuprer l'anne actuelle
annee_actuelle = datetime.now().year

# Calculer l'ge en soustrayant l'anne de naissance de l'anne actuelle
labonnedata2_df['age'] = annee_actuelle - labonnedata2_df['birth']

# Afficher le DataFrame avec la nouvelle colonne 'age'
labonnedata2_df
```

```python
import matplotlib.pyplot as plt
# Crer des catgories d'ge par regroupement par intervalles de 5 ans
labonnedata2_df['grouped_age'] = pd.cut(labonnedata2_df['age'], bins=range(labonnedata2_df['age'].min(), labonnedata2_df['age'].max()+1, 10))
# Crer un DataFrame avec le nombre de clients par sexe et catgorie
data_counts = labonnedata2_df.groupby(['grouped_age', 'categ']).size().unstack()


# Crer le graphique  barres
data_counts.plot(kind='bar', stacked=True)

# Ajouter des tiquettes et un titre
plt.title("Rpartition des achats par catgories et par tranches d'ge")
plt.xlabel("Tranches d'ge")
plt.ylabel('Nombres de clients')
plt.xticks(rotation=45, ha='right')

# Afficher le graphique
plt.legend(title='Catgories')
plt.tight_layout()
plt.show()
```

```python
# Convertir la colonne 'date' en type datetime si ce n'est pas dj fait
labonnedata2_df['date'] = pd.to_datetime(labonnedata2_df['date'])

# Ajouter une colonne 'jour' pour extraire la date  partir de 'date'
labonnedata2_df['jour'] = labonnedata2_df['date'].dt.date

# Compter le nombre de session_id par jour
sessions_par_jour = labonnedata2_df.groupby('jour')['session_id'].count()

# Crer un graphique en courbe pour visualiser les donnes
plt.figure(figsize=(10, 6))
sessions_par_jour.plot(kind='line', color='skyblue', marker='o', linestyle='-')
plt.title('Nombre de sessions par jour')
plt.xlabel('Jour')
plt.ylabel('Nombre de sessions')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
